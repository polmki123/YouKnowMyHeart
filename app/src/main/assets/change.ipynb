```python
import os
from PIL import Image, ImageFilter, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
img = os.listdir('FoodImage')
for _i in img :
    print(_i)
    with Image.open( os.path.join('FoodImage', _i) ) as im :
        # changeImg = Image.new("RGB", im.size, (255,255,255))
        # changeImg.paste(im, mask=im.split()[3])
        try : 
            changeImg = im
            changeImg.save(os.path.join('newImage', _i), 'jpeg', quality = 25)
        except :
            changeImg = Image.new("RGB", im.size, (255,255,255))
            changeImg.paste(im, mask=im.split()[3])
            changeImg.save(os.path.join('newImage', _i), 'jpeg', quality = 25)
```
